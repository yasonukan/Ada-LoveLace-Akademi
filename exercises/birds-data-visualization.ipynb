```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
birds = pd.read_csv('data/birds.csv')
```

```python
birds.head()
```

```python
birds.info()
```

```python
birds.plot.scatter( x='MaxLength', y='Order', figsize=(12,8))

plt.title('Max Length per Order')
plt.ylabel('Order')
plt.xlabel('Max Length')
#This gives an overview of the general distribution of body length per bird Order
```

```python
birds['MaxBodyMass'].plot(kind = 'hist', bins = 10, figsize = (5,5))
plt.show()
```

```python
birds['MaxLength'].plot(kind = 'hist', bins = 10, figsize = (5,5))
plt.show()
```

```python
filteredBirds = birds[(birds['MaxBodyMass'] > 1) & (birds['MaxBodyMass'] < 100)]      
check = birds[(birds['MaxLength'] > 1) & (birds['MaxLength'] < 100)]      

```

```python
filteredBirds.head()
```

```python
filteredBirds['MaxBodyMass'].plot(kind = 'hist',bins = 40,figsize = (8,8))
plt.show()  
```

```python
check['MaxLength'].plot(kind = 'hist',bins = 40,figsize = (8,8))
plt.show()  
```

```python
sns.kdeplot(
   data=birds, x="MaxBodyMass", hue="Order",
   fill=True, common_norm=False, palette="pastel",
   alpha=.5, linewidth=0, warn_singular=False,
)
```

```python
sns.kdeplot(
   data=filteredBirds, x="MaxBodyMass", hue="Order",
   fill=True, common_norm=False, palette="pastel",
   alpha=.5, linewidth=0, warn_singular=False,
)
```

```python
sns.kdeplot(
   data=check, x="MaxLength", hue="Order",
   fill=True, common_norm=False, palette="pastel",
   alpha=.5, linewidth=0, warn_singular=False,
)
```

---

```python
# MaxBodyMass - MaxWingSpan 

sns.regplot(
    x=filteredBirds['MaxBodyMass'],
    y=filteredBirds['MaxWingspan'],
    marker= "+",
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5 
    },
    scatter_kws={
        "color":"darkblue",
        "s":50
    }
    )
```

```python
# MaxLength - MaxWingSpan 


sns.regplot(
    x=check['MaxLength'],
    y=check['MaxWingspan'],
    marker= "+",
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5 
    },
    scatter_kws={
        "color":"darkblue",
        "s":50
    }
    )
```

```python
sns.lmplot( 
    x='MaxBodyMass',
    y='MaxLength',
    data=birds,
    hue="Order"
)
```

```python
sns.lmplot( 
    x='MaxBodyMass',
    y='MaxLength',
    data=filteredBirds, #MaxBodyMass filtered
    hue="Order"
)
```

```python
sns.lmplot( 
    x='MaxBodyMass',
    y='MaxLength',
    data=check,  #MaxLength filtered 
    hue="Order"
)
```
