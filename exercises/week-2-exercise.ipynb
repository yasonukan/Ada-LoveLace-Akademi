```python
# You can find covid related data under project's data folder 
base_url = "../data/covid/"
infected_dataset_url = base_url + "time_series_covid19_confirmed_global.csv"
recovered_dataset_url = base_url + "time_series_covid19_recovered_global.csv"
deaths_dataset_url = base_url + "time_series_covid19_deaths_global.csv"
```

#### Solution:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set style & figures inline
sns.set()
%matplotlib inline
```

```python
infected_data = pd.read_csv(infected_dataset_url)
recovered_data = pd.read_csv(recovered_dataset_url)
deaths_data = pd.read_csv(deaths_dataset_url)
```

### Confirmed Data

```python
# check it out
infected_data.head()
```

```python
# Printing a summary of the Infected Data.
# This summary includes information such as the number of rows,
# the number of columns, and the data type of each column.
infected_data.info()
```

```python
# Calling the 'describe' method on the 'infected_data' DataFrame.
# This method returns a summary of the central tendency, dispersion and shape of the dataset's distribution.
infected_data.describe()
```

```python
infected_data.shape
```

```python
# Group by region
# Filtering out the relevant data columns (i.e., Country/Region, Lat, Long, and the daily counts)
# Grouping the data by the region (Country/Region) and calculating the cumulative daily counts for each region
infected_data_days = infected_data.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)

# Printing the first few rows of the grouped data

infected_data_days.head()
```

```python
df = infected_data_days.transpose()
df.head()
```

```python
# Melt the DataFrame to transform it into a long format

data = infected_data_days.reset_index().melt(id_vars='Country/Region', var_name='date')
data.head()
```

```python
# The data frame contains two columns: 'date' and 'value'. 
# We want to update the 'value' column where the 'value' is less than 1.

# Here, we are using the 'loc' function to locate the rows in the data frame 
# where the 'value' is less than 1. The second argument, 'value', indicates that 
# we want to select the 'value' column. The first argument, the boolean expression, 
# returns True for rows where the 'value' is less than 1, and False otherwise.

# By setting the values to None, we effectively convert the values to NaN.

data.loc[(data.value < 1),'value'] = None
data.head()
```

```python
# Pivot the data to a wide format with countries as columns and dates as index
df = data.pivot(index='date', columns='Country/Region', values='value')

# Display the last few rows of the resulting DataFrame
df.tail()
```

```python
# Set index as DateTimeIndex

# Create a DatetimeIndex object using the 'index' of 'df'
datetime_index = pd.DatetimeIndex(df.index)

# Set the DatetimeIndex as the index of 'df' using the 'inplace' parameter to avoid creating a new DataFrame
df.set_index(datetime_index, inplace=True)

df.head()
```

```python
# Check out index
df.index
```

```python
# Plot time series of several countries of interest
poi = ['Turkey', 'US', 'China', 'France', 'Australia']
df[poi].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20);
plt.ylabel('Confirmed patients count', fontsize=20);
plt.title('Confirmed Patients Time Series', fontsize=20);
```

```python
# Plot time series of several countries of interest
poi = ['Turkey', 'US', 'China', 'France', 'Australia']
df[poi].plot(figsize=(20,10), linewidth=5, fontsize=20, logy=True)
plt.xlabel('Date', fontsize=20);
plt.ylabel('Confirmed Patients Logarithmic count', fontsize=20);
plt.title('Confirmed Patients Logarithmic Time Series', fontsize=20);
```

### Deaths

```python
deaths_data.head()
```

```python
# Reshape the deaths_data to get a date-based summary
deaths_data_days = deaths_data.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)
df = deaths_data_days.transpose()

# Prepare the data for the time series visualization
data = deaths_data_days.reset_index().melt(id_vars='Country/Region', var_name='date')

# Filter out death counts that are less than 25 to clean up the visualization
data.loc[(data.value < 25),'value'] = None

# Create a new dataframe with dates as index and countries as columns
df = data.pivot(index='date', columns='Country/Region', values='value')

# Convert the date index to datetime
datetime_index = pd.DatetimeIndex(df.index)
df.set_index(datetime_index, inplace=True)
```

```python
df.tail()
```

```python
# Plot time series of several countries of interest
poi = ['Turkey', 'US', 'China', 'France', 'Australia']
df[poi].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20);
plt.ylabel('Deaths Patients count', fontsize=20);
plt.title('Deaths Patients Time Series', fontsize=20);
```

```python
# Using dropna method to remove all null columns from the DataFrame
df.dropna(axis=1, how='all', inplace=True)

df.head()
```

```python
# First, sort the DataFrame based on its index using the sort_index() method
df = df.sort_index()

# Then, reset the index of the DataFrame using the reset_index() method
# This method returns a new DataFrame with the same columns and index
# It also allows us to specify whether the index should be added as a column to the DataFrame
df1 = df.reset_index().drop(['date'], axis=1)

# Finally, print the first few rows of the modified DataFrame using the head() method
df1.head()
```

```python
# Loop through each column in the DataFrame df1
for col in df1.columns:
    
    # Print the column name and the index of the first valid value in the column
    print(col, df1[col].first_valid_index())
    
    # Shift the values in the column to align with the first valid value
    # This operation creates a new DataFrame where the values in the column are shifted by -df1[col].first_valid_index() positions
    df1[col] = df1[col].shift(-df1[col].first_valid_index())
```

```python
# We use the 'apply' function to apply a function along each column/row of the dataframe
df2 = df1.apply(lambda x: x.shift(-x.first_valid_index()))

# Let's break down the lambda function
# 1. 'x' is an element in the dataframe 'df1'
# 2. 'x.shift(-x.first_valid_index())' is the equivalent of shifting each element in 'x' by the first valid index of 'x' in reverse
# 3. The lambda function applies this shift operation to each element in 'df1' and stores the result in 'df2'
```

```python
# Plot time series of several countries of interest
# Create a figure of size 20x10 with 5 lines
df2.plot(figsize=(20,10), linewidth=5, fontsize=20)

# Set labels and title
plt.xlabel('Days', fontsize=20);
plt.ylabel('Deaths Patients count', fontsize=20);
plt.title('Deaths Patients Time Series', fontsize=20);
```

```python
# Plot time series of several countries of interest
df2.plot(figsize=(20,10), linewidth=5, fontsize=20, logy=True)
plt.xlabel('Days', fontsize=20);
plt.ylabel('Deaths Patients Logarithmic count', fontsize=20);
plt.title('Deaths Patients Logarithmic Time Series', fontsize=20);
```
